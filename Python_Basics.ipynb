1. What is Python, and why is it popular ?

2. What is an interpreter in Python?

3. What are pre-defined keywords in Python ?
  - Pre-defined keywords are the reserved keywords in python that have special meanings and purposes.
4. Can keywords be used as variable names ?
  - No keywords can never be used as variables names in python because they are already reserved by the python for some general or specific purpose.
5. What is mutability in Python ?
  - Mutablility in python means that the whatever the content in the object can be updated or changed after creation.      
6. Why are lists mutable, but tuples are immutable ?
- This is because the lists are designed for dynamic data handling as we can add,delete and update the data that's why mutable while the tuple are designed to store and handle fixed data we cannot update the content once created that's why immutable.
7. What is the difference between “==” and “is” operators in Python ?
  - The "==" operator is used to compare the data of two or more objects.
  - On other hand, "is" operator is used to compare the identities, it is used to checks if two variables refer to the same object in memory.
8. What are logical operators in Python ?
  - Logical operators are used to combine conditional statements:
    - and → True if both conditions are True.
    - or → True if at least one condition is True.
    - not → Reverses the logical value (True → False, False → True)
9. What is type casting in Python ?
  - Type casting in python refers to the conversion of a variable from one data type to other data type.
  - For Example: a = "22" , b = 20 now if we want to sum up these two variables by simply writing a + b without type casting "a" into "int" data type , it returns an error. But if we type cast "a" into "int" data type as int(a) + b , then it return 42 as an answer.
10. What is the difference between implicit and explicit type casting ?
  - Implicit casting: is when python automatically converts data types during operations.
  - Explicit casting: is when user manually convert data types using functions like int(), float(), str(), etc.
11. What is the purpose of conditional statements in Python ?
  - Conditional statements control the flow of code execution by executing only those statements for which the certain conditions returns true.
12. How does the elif statement work ?
  - elif statement is used in nested conditional checks where we are checking for more than one or two conditions .
13. What is the difference between for and while loops ?
  - for loops is used when the number of iterations are known.
  - while loops are used when the number of iterations are unknown.
14. Describe a scenario where a while loop is more suitable than a for loop.
  - the use of while loop is more suitable when we dont know how many times we have to run the loop means the number of iterations are unknown.
  - For example : if we checking for whether the user has entered the correct password or not, in this case the use of while loops is better.

```python
# Question 4 : Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.
list =[1,2,4,5,3]
print(f"This is our list : {list}")
print(f"Before updation the element at 0 index will be {list[0]}")
list[0] = 6
print(f"After updation the element at 0 index will be {list[0]}")
# till here every thing runs fine because list in python allows the user to change or update the content within it.
tuple = (1,3,5,2,4)
print(f"This is a tuple : {tuple}")
print(f"Before updation the element at 0 index will be {tuple[0]}")
try:
    tuple[1] = 6  # This will raise an error
except TypeError as e:
    print("Error when modifying tuple:", e)
#the last line will throw an error because tuple is an object and we cannot change or assign any value to any element or any index in the tuple.
```

```python
# Question 6 : Write a function to demonstrate the behavior of mutable and immutable arguments ?

def modify_list(lst):
    lst.append(100)  # modifies the original list

def modify_number(num):
    num += 10  # creates a new number (doesn't affect original)

# Test
my_list = [1, 2, 3]
my_num = 5

modify_list(my_list)
modify_number(my_num)

print("After modifying list:", my_list)    # list changed
print("After modifying number:", my_num)   # number unchanged
```

```python
# Question 14 : Write a program to reverse a string using a while loop

name = input("Enter your name: ")
rev = ""
i = len(name) - 1;
while i >= 0 :
  rev += name[i]
  i -= 1
print(rev)
```

```python
# Question 15 :  Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial of", num, "is", fact)
```
